# Seaborn Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Orders.csv')
df
```

```python
sns.histplot(df['Quantity'], bins = 5)
```

```python
sns.histplot(df['Quantity'], bins = 8)
```

```python
sns.barplot(x = 'Category', y = 'Unit Price', data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
myGraph = sns.scatterplot(x = 'Discount', y = 'Total Cost', data = df, size = 'Quantity', hue = 'Category')
myGraph.legend(bbox_to_anchor = (1, 1), loc = 'upper left')
```
