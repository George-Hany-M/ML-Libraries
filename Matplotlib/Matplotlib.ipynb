# import Libraries

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

# Plot Data

```python
Days = ['Sat', 'Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri']
Temp_Cairo = [22.5, 22.7, 25, 27.3, 26.1, 22.4, 20]
Temp_Fayoum = [30.2, 29, 25, 33.3, 39.7, 35, 28]

DF_Cairo = pd.DataFrame({'Days' : Days, 'Temp_Cairo' : Temp_Cairo})
DF_Fayoum = pd.DataFrame({'Days' : Days, 'Temp_Fayoum' : Temp_Fayoum})
```

```python
DF_Cairo
```

```python
DF_Fayoum
```

# Line Plots

```python
plt.plot(DF_Cairo['Days'], DF_Cairo['Temp_Cairo'])
plt.show()
```

```python
plt.plot(DF_Fayoum['Days'], DF_Fayoum['Temp_Fayoum'])
plt.show()
```

# Scatter Plots

```python
plt.scatter(DF_Cairo['Days'], DF_Cairo['Temp_Cairo'])
plt.show()
```

```python
plt.scatter(DF_Fayoum['Days'], DF_Fayoum['Temp_Fayoum'])
plt.show()
```

# Bar Charts

```python
plt.bar(DF_Cairo['Days'], DF_Cairo['Temp_Cairo'])
plt.show()
```

```python
plt.bar(DF_Fayoum['Days'], DF_Fayoum['Temp_Fayoum'])
plt.show()
```

# Customizing Plots

```python
plt.plot(DF_Cairo['Days'], DF_Cairo['Temp_Cairo'])

# x-axis & y-axis
plt.xlabel('Days of the week')
plt.ylabel('Temperatures in Cel.')

# title
plt.title('Temperatures Degrees in Cairo')

plt.show()
```

```python
plt.plot(DF_Cairo['Days'], DF_Cairo['Temp_Cairo'], label = 'Cairo')
plt.plot(DF_Fayoum['Days'], DF_Fayoum['Temp_Fayoum'], label = 'Fayoum', c = 'r', alpha = 0.5)

# x-axis & y-axis
plt.xlabel('Days of the week')
plt.ylabel('Temperatures in Cel.')

# title
plt.title('Temperatures Degrees in Cairo')

# plot titles
plt.legend()

plt.show()
```
